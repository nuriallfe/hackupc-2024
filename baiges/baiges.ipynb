# DISTANCES FROM COORDENATES

```python
import math

def haversine_distance(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))

    # Radius of Earth in kilometers. Use 3956 for miles
    r = 6371

    # Calculate the result
    return c * r

# Example usage
latitude1 = 52.2296756
longitude1 = 21.0122287
latitude2 = 41.8919300
longitude2 = 12.5113300

distance = haversine_distance(latitude1, longitude1, latitude2, longitude2)
print(f"The distance between the two points is {distance:.2f} kilometers.")
```

# BAIGES GET IMAGES

```python
import requests
import os

def fetch_images(api_key, text_search, num_images=10):
    """
    Fetches images from Flickr API based on a text search, ensuring it gathers a specified number of valid image URLs.

    Args:
        api_key (str): Flickr API key
        text_search (str): Text to search for
        num_images (int): Desired number of images to fetch

    Returns:
        list: List of image URLs
    """
    url = 'https://api.flickr.com/services/rest/'
    images = []
    page = 1

    while len(images) < num_images:
        params = {
            'method': 'flickr.photos.search',
            'api_key': api_key,
            'text': text_search,
            'sort': 'relevance',
            'media': 'photos',
            'safe_search': 1,
            'extras': 'url_l',  # 'url_l' is more likely to be available
            'format': 'json',
            'nojsoncallback': 1,
            'per_page': 100,  # Fetch more photos per request
            'page': page
        }

        try:
            response = requests.get(url, params=params)
            response.raise_for_status()
            photos = response.json()['photos']['photo']
            for photo in photos:
                if 'url_l' in photo and len(images) < num_images:
                    images.append(photo['url_l'])
            page += 1
        except requests.RequestException as e:
            print(f"Error fetching data: {e}")
            break

    return images

def save_images(images, directory, text_search):
    """
    Saves a list of images to a directory with filenames based on the search term.

    Args:
        images (list): List of image URLs
        directory (str): Directory to save the images
        text_search (str): Text search term used to name the files

    Returns:
        None
    """
    if not os.path.exists(directory):
        os.makedirs(directory)
    for i, image_url in enumerate(images):
        filename = f'{text_search.replace(" ", "_")}_{i+1}.jpg'  # Replace spaces with underscores and append the index
        try:
            response = requests.get(image_url)
            response.raise_for_status()
            with open(os.path.join(directory, filename), 'wb') as f:
                f.write(response.content)
            print(f'Image {filename} saved successfully.')
        except requests.RequestException as e:
            print(f"Error downloading image {filename}: {e}")
```

# IMAGES EMBEDDINGS

```python
import torch
from torchvision import models, transforms
from torch.nn.functional import cosine_similarity
from PIL import Image
import glob
import re


def load_image(image_path):
    """
    Load an image from a file and apply the necessary transformations

    Args:
        image_path (str): Path to the image file

    Returns:
        torch.Tensor: Transformed image tensor
    """
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ])
    image = Image.open(image_path).convert('RGB')  # Convertir la imagen a RGB
    image = transform(image).unsqueeze(0)  # Añade una dimensión al principio
    return image


def get_embedding(model, image_tensor):
    """
    Get the embedding of an image from a pre-trained model

    Args:
        model (torch.nn.Module): Pre-trained model
        image_tensor (torch.Tensor): Image tensor

    Returns:
        torch.Tensor: Image embedding
    """
    with torch.no_grad():
        embedding = model(image_tensor)
    return embedding

def find_most_similar(reference_embedding, embeddings):
    """
    Find the index of the most similar image to a reference embedding

    Args:
        reference_embedding (torch.Tensor): Reference embedding
        embeddings (list): List of embeddings

    Returns:
        int: Index of the most similar image
    """
    similarities = [cosine_similarity(reference_embedding, emb, dim=1) for emb in embeddings]

    # Find the top 3 most similar images
    most_similar_indices = sorted(range(len(similarities)), key=lambda i: similarities[i], reverse=True)[:3]
    return most_similar_indices

# Load a pre-trained ResNet model
model = models.resnet18(pretrained=True)
model.eval()

# Extract the embeddings of the images
reference_image_path = "vesubio.jpg"
image_list_paths = glob.glob("downloaded_images/*.jpg")

reference_image = load_image(reference_image_path)
reference_embedding = get_embedding(model, reference_image)

image_embeddings = [get_embedding(model, load_image(img_path)) for img_path in image_list_paths]

# Find the most similar image to the reference image
most_similar_index = find_most_similar(reference_embedding, image_embeddings)

# Find the most similar real landmarks

def clean_image_name(file_path):
    # Delete the directory path and the file extension
    base_name = os.path.basename(file_path).replace('.jpg', '')
    
    # Delete any numbers and underscores at the end of the file name
    clean_name = re.sub(r'_[0-9]+$', '', base_name)
    
    # Replace any remaining underscores with spaces
    clean_name = clean_name.replace('_', ' ')
    
    return clean_name

filenames = [clean_image_name(image_list_paths[i]) for i in most_similar_index]

print("The most similar images are:", filenames)

print("The most similar image is at index:", most_similar_index[0], "Path:", image_list_paths[most_similar_index[0]])
print("The second most similar image is at index:", most_similar_index[1], "Path:", image_list_paths[most_similar_index[1]])
print("The third most similar image is at index:", most_similar_index[2], "Path:", image_list_paths[most_similar_index[2]])


# Visualize the top 3 most similar images

import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 4, figsize=(15, 5))

# Load and plot the reference image
reference_image = Image.open(reference_image_path)
axs[0].imshow(reference_image)
axs[0].set_title("Reference Image")
axs[0].axis('off')

# Load and plot the most similar image
most_similar_image = Image.open(image_list_paths[most_similar_index[0]])
axs[1].imshow(most_similar_image)
axs[1].set_title(f'Most Similar Image: {most_similar_index[0]}')
axs[1].axis('off')

# Load and plot the second most similar image
second_most_similar_image = Image.open(image_list_paths[most_similar_index[1]])
axs[2].imshow(second_most_similar_image)
axs[2].set_title(f'Secong Most Similar Image: {most_similar_index[1]}')
axs[2].axis('off')

# Load and plot the third most similar image
third_most_similar_image = Image.open(image_list_paths[most_similar_index[2]])
axs[3].imshow(third_most_similar_image)
axs[3].set_title(f'Third Most Similar Image: {most_similar_index[2]}')
axs[3].axis('off')
plt.show()
```
